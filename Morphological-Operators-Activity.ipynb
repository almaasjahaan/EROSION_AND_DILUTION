# Morphological Operators

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Replace 'ABCDE' with your name (only the first 5 characters)
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ALLU',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

## Erosion

Erodes away boundaries of foreground objects. Works best when foreground is light color (preferrably white) and background is dark.

```python
# Create a 5x5 kernel filled with ones and display it using dtype=np.uint8
```

```python
kernel = np.ones((5,5), dtype=np.uint8)
```

```python
kernel
```

```python
# Apply erosion on the image using the specified kernel with one iteration and store the result in erosion1 using the cv2.erode() function.
```

```python
erosion1 = cv2.erode(img, kernel, iterations=1)
```

```python
display_img(erosion1)
```

```python

# start from the main image already loaded earlier
black_noise_img = img.copy()

# add random black noise (scattered black pixels)
prob = 0.10  # 10% of pixels are turned to black – increase this for more scatter
rand = np.random.rand(*black_noise_img.shape[:2])

black_noise_img[rand < prob] = 0

display_img(black_noise_img)
```

```python
import numpy as np

# start from the main image already loaded earlier
black_noise_img = img.copy()

# convert to grayscale so we can work on single channel
gray = cv2.cvtColor(black_noise_img, cv2.COLOR_BGR2GRAY)

# create an all‑black image
noisy = np.zeros_like(gray)

# create a mask where the original text is (bright region)
_, text_mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# add white scatter only inside the text region
prob = 0.6  # increase for denser scatter (0.6–0.8 looks like your example)
rand = np.random.rand(*gray.shape)

# where mask is text and random < prob → set to white
noisy[(text_mask == 255) & (rand < prob)] = 255

black_noise_img = noisy
display_img(black_noise_img)
```

```python
black_noise_img[black_noise_img==-255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```

## Morphological Gradient

Difference between dilation and erosion of an image.

```python
img = load_img()
```

```python
display_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
